##### Convert the continous image to grayscale

```python
vid = cv2.VideoCapture(0)
while True:
    flag , img = vid.read()
    if flag is True:
        img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        cv2.imshow('Live',img_gray[::-1,:])
        key = cv2.waitKey(1)
        if key == ord('a'):
            break
    else:
        print("Not Running")
        break
vid.release()
cv2.destroyAllWindows()
```

##### Binary Image

```python
vid = cv2.VideoCapture(0)
while True:
    flag , img = vid.read()
    if flag is True:
        img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        _,img_binary = cv2.threshold(img_gray,50,255,cv2.THRESH_BINARY)
        cv2.imshow('Live',img_binary)
        key = cv2.waitKey(1)
        if key == ord('a'):
            break
    else:
        print("Not Running")
        break
vid.release()
cv2.destroyAllWindows()
```

##### Reactangle highlightted image

```python
vid = cv2.VideoCapture(0)
while True:
    flag , img = vid.read()
    if flag is True:
        x1,y1,w,h = (800,400,700,200)
        img_crop = img[y1:y1+h,x1:x1+w,:]
        cv2.rectangle(
            img,pt1 = (x1,y1), pt2 = (x1+w,y1+h),
                     color = (0,0,255),thickness = 10
                     )
        cv2.imshow('Live',img)
        key = cv2.waitKey(1)
        if key == ord('a'):
            break
    else:
        print("Not Running")
        break
vid.release()
cv2.destroyAllWindows()
```

### haarcascade open cv github
        face detection

```python
vid = cv2.VideoCapture(0)

fd = cv2.CascadeClassifier(
     cv2.data.haarcascades+
     'haarcascade_frontalface_default.xml')
while True:
    flag , img = vid.read()
    if flag is True:
        img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        
#         x1,y1,w,h = (800,400,700,200)
        faces = fd.detectMultiScale(img_gray, 1.1, 5)
        for x1,y1,w,h in faces:
            
            img_crop = img[y1:y1+h,x1:x1+w,:]
            cv2.rectangle(
                img,pt1 = (x1,y1), pt2 = (x1+w,y1+h),
                         color = (0,0,255),thickness = 10
                         )
        cv2.imshow('Live',img)
        key = cv2.waitKey(1)
        if key == ord('a'):
            break
    else:
        print("Not Running")
        break
vid.release()
cv2.destroyAllWindows()
```
