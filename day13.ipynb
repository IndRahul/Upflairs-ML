```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage.transform import rotate
```

```python
img1 = cv2.imread("file1.jpg")
img2 = cv2.imread('file2.png')
img2 = img2[0:0+122,0:0+183,:]
```

```python
img1.shape,img2.shape
```

```python
def imshow(img):
    cv2.imshow('Preview',img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
```

```python
# imshow(img1)
imshow(img2)
```

```python
def showimg(img):
    plt.imshow(img[:,:,::-1])
    plt.show()
```

```python
showimg(img1)
```

```python
showimg(img2)
```

```python
add = cv2.add(img1,img2)
showimg(add)
```

```python
sub = cv2.subtract(img1,img2)
showimg(sub)
```

```python
cont_img1 = cv2.multiply(img1, 2 * np.ones(img1.shape).astype('uint8'))
showimg(cont_img1)
```

Brightness decrease

```python
bright_img1 = cv2.subtract(img1, 20 * np.ones(img1.shape).astype('uint8'))
showimg(bright_img1)
```

FUnction for controlling brightness and contrast directly by the function 

```python
bright_img2 = cv2.convertScaleAbs(img2,225,199)
showimg(bright_img2)
```

Interpolation technique --> removing some elements from many   10X10 --> 5X5


Extrapolation --> adding some elements from low to many   5X5 --> 10X10


```python
resize_img2 = cv2.resize(img2,(200,800),interpolation=cv2.INTER_CUBIC)
resize_img2.shape
```

```python
showimg(resize_img2)
```

```python
resize_img2 = cv2.resize(img2,None,fx=5,fy=3,interpolation=cv2.INTER_CUBIC)
resize_img2.shape
```

```python
showimg(resize_img2)
```
